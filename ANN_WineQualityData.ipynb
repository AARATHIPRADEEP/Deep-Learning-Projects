```python
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/Dataset1/winequality-red.csv')
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['quality']=le.fit_transform(df['quality'])
```

```python
df['quality'].value_counts()
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
from keras.utils import to_categorical
y=to_categorical(y)
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=1)
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
ann=Sequential()
ann.add(Dense(units=11,activation='relu'))
ann.add(Dense(units=20,activation='relu'))
ann.add(Dense(units=6,activation='softmax'))
```

```python
ann.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(X_train,y_train,batch_size=10,epochs=50)
```

```python
loss,accuracy=ann.evaluate(X_test,y_test)
print('Loss=',loss)
print('Accuracy=',accuracy)
```

```python
y_pred=ann.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay

print(classification_report(y_test.argmax(axis=1),y_pred.argmax(axis=1)))
print(ConfusionMatrixDisplay.from_predictions(y_test.argmax(axis=1),y_pred.argmax(axis=1)))
```
